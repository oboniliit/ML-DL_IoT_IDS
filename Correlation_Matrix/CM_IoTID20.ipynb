```
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt # plotting
import numpy as np # linear algebra
import os # accessing directory structure
import pandas as pd 
```

```
nRowsRead = 200 # specify 'None' if want to read whole file
# Preprossed_data.csv may have more rows in reality, but we are only loading/previewing the first 1000 rows
df1 = pd.read_csv('/content/drive/MyDrive/Code/dataset/IoT Intrusion Dataset 2020.csv', delimiter=',', nrows = nRowsRead)
df1.dataframeName = 'IoTID20.csv'
nRow, nCol = df1.shape
print(f'There are {nRow} rows and {nCol} columns')
```
